# 2.设置x的刻度 xticks,加上这个就等于把x能渠道得值都显示出来

```python
from matplotlib import pyplot as plt
x= range(2,26,2)
y=[15,13,14.5,17,20,25,26,26,24,22,18,15]
plt.figure(figsize=(20,8),dpi=80)
_xtick_labels = [i/2 for i in range(4,49)]
#plt.xticks(range(2,25))
plt.xticks(_xtick_labels)
plt.yticks(range(min(y),max(y)+1))

plt.plot(x,y)
plt.show()#看效果 
```

# 3.实例，绘制气温

```python
from matplotlib import pyplot as plt
import random
x = range(0,120)
y = [random.randint(20,35) for i in range(120)]
plt.figure(figsize=(20,8),dpi = 80)
plt.plot(x,y)
plt.show()
```

# 4.绘制气温修改，调整刻度
# plt.legent()设置标签，loc = ‘注明标签位置’

```python
from matplotlib import pyplot as plt
from matplotlib import font_manager
import random
my_font = font_manager.FontProperties(fname =  'C:\Windows\Fonts\STSONG.TTF',size = 18)
x = range(0,121)
y = [random.randint(20,35) for i in range(121)]
plt.figure(figsize=(20,8),dpi = 80)
plt.plot(x,y,color = 'red',label='气温',linestyle=':')
_x = list(x)[::10]
_xtick_labels = ['T{}'.format(i) for i in _x]
plt.xticks(_x,_xtick_labels,fontproperties = my_font,rotation = 45)
plt.legend(prop = my_font,loc='upper left')
plt.xlabel("时间",fontproperties = my_font)
plt.ylabel("温度",fontproperties = my_font)
plt.title("气温变化情况",fontproperties = my_font)

plt.show()
```

# 上面这个图形，plt.xlabel表示设置x轴是什么，plt.ylabel表示设置y轴是什么，同时注意把字体设置进去

# 5.预测女朋友的走势

```python
from matplotlib import pyplot as plt
from matplotlib import font_manager
x = range(11,31)
y = [1,0,1,1,2,4,3,2,3,4,4,5,6,5,4,3,3,1,1,1]
s = [2,3,1,2,5,6,8,1,0,0,3,6,2,1,2,5,2,1,1,1]
plt.figure(figsize=(20,8),dpi = 80)
my_font = font_manager.FontProperties(fname = 'C:\Windows\Fonts\STSONG.TTF',size = 18)
_xtick_labels = ["{}岁".format(i) for i in x]
plt.xticks(x,_xtick_labels,fontproperties = my_font,rotation = 45)
plt.xlabel('年龄',fontproperties = my_font)
plt.ylabel('数量',fontproperties = my_font)
plt.plot(x,y,color = 'red',label ='小明',linestyle = '-')
plt.plot(x,s,color = 'b',label ='小蓝',linestyle = ':')
plt.legend(prop = my_font,loc='upper left')
plt.grid()


plt.show()
```

# 7.绘制散点图，气温变化规律

```python
from matplotlib import pyplot as plt
from matplotlib import font_manager
import random
x1 = range(1,31)
x2 = range(51,81)
y1 = [11,17,16,11,12,11,12,6,6,7,8,9,12,15,14,17,18,21,16,17,20,14,15,15,15,19,21,22,22,22]
y2= [26,26,28,19,21,17,16,19,18,20,20,19,22,23,17,20,21,20,22,15,11,15,5,13,17,10,11,13,12,13]
plt.figure(figsize=(20,8),dpi = 80)
#调整x轴的刻度
my_font = font_manager.FontProperties(fname="C:\Windows\Fonts\STSONG.TTF",size = 18)
_x = list(x1)+list(x2)
_xtick_labels = ["3月{}日".format(i) for i in x1]
_xtick_labels += ["10月{}日".format(i-50) for i in x2]

plt.xticks(_x[::3],_xtick_labels[::3],fontproperties = my_font,rotation = 45)
plt.scatter(x1,y1,label='三月份',color = 'k')
plt.scatter(x2,y2,label='十月份',color = 'r')
plt.xlabel('时间',fontproperties = my_font)
plt.ylabel('温度℃',fontproperties = my_font,rotation=90)
plt.title('天气气温变化',fontproperties = my_font,color = 'r',size = 25)

plt.legend(prop = my_font,loc = 'upper left')
plt.show()
```
